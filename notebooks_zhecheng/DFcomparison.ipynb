```python
import sys
import os

sys.path.append("..")

import pandas as pd
import numpy as np

import random
import itertools
from sklearn import metrics


from myutils import number_split, create_mix
```

```python
np.random.seed(12)

df = pd.DataFrame({"z": np.random.binomial(1, 0.2, 2000),
                   "y": np.random.binomial(1, 0.5, 2000)
                  })

df0 = df.query("z == 0").reset_index(drop=True)
df1 = df.query("z == 1").reset_index(drop=True)
```

```python
df.groupby(['z','y']).size()
```

```python
## create dataset
```

```python
n_test = 150
train_test_ratio = 4


valid_full_settings = []
for combination in itertools.product([0.1, 0.5], 
                                    [0.2, 0.6], 
                                     np.arange(0.1, 0.999, 0.1),
                                     np.arange(0, 5, 0.05)
                                    ):
    

    number_setting = number_split(p_pos_train_z0=combination[0], 
                           p_pos_train_z1 = combination[1], 
                           p_mix_z1 = combination[2], alpha_test = combination[3],
                           train_test_ratio = train_test_ratio, 
                           n_test=n_test)

    if number_setting is not None:
        valid_full_settings.append(number_setting)
    else:
        print(combination)
```

```python
tmp = valid_full_settings[0].copy()
df_list = {}
params = tmp['mix_param_dict']
del tmp['mix_param_dict']
tmp.update(params)

allkeys = tmp.keys()
        
```

```python
allkeys
```

```python
for k in allkeys:
    df_list[k] = []
   
            

for i, st in enumerate(valid_full_settings):
    
    dfs = create_mix(df1 = df1, df0 = df0, target='y', setting= st)
    if dfs is not None:
        params = dfs['setting']['mix_param_dict']
        dfs['setting'].update(params)
        for k in df_list.keys():
            if k != 'mix_param_dict':
                df_list[k].append(dfs['setting'][k])
```

```python
final_df = pd.DataFrame.from_dict(df_list)
```

```python
final_df = final_df.iloc[:,2:]
```
